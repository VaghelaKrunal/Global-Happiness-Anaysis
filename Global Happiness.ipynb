```
df_2015_clean = df_2015[['Happiness Score', 'Economy (GDP per Capita)', 'Year']].rename(columns={'Economy (GDP per Capita)': 'GDP per Capita'})
df_2016_clean = df_2016[['Happiness Score', 'Economy (GDP per Capita)', 'Year']].rename(columns={'Economy (GDP per Capita)': 'GDP per Capita'})
df_2017_clean = df_2017[['Happiness Score', 'Economy..GDP.per.Capita.', 'Year']].rename(columns={'Economy..GDP.per.Capita.': 'GDP per Capita'})
df_2018_clean = df_2018[['Happiness Score', 'GDP per capita', 'Year']].rename(columns={'GDP per capita': 'GDP per Capita'})

combined_df = pd.concat([df_2015_clean, df_2016_clean, df_2017_clean, df_2018_clean], ignore_index=True)

annual_averages = combined_df.groupby('Year')[['Happiness Score', 'GDP per Capita']].mean().reset_index()

data_2015 = annual_averages[annual_averages['Year'] == 2015]
h_avg = data_2015['Happiness Score'].values[0]
gdp_avg = data_2015['GDP per Capita'].values[0]

plt.figure(figsize=(4, 6))
plt.bar(["Happiness Score", "GDP per Capita"], [h_avg, gdp_avg], color=["skyblue","salmon"])
plt.title("Average Happiness Score vs GDP per Capita (2015)")
plt.ylabel("Average Value")
plt.show()
```

```
data_2016 = annual_averages[annual_averages['Year'] == 2016]
h_avg = data_2016['Happiness Score'].values[0]
gdp_avg = data_2016['GDP per Capita'].values[0]

plt.figure(figsize=(4, 6))
plt.bar(["Happiness Score", "GDP per Capita"], [h_avg, gdp_avg], color=["skyblue","salmon"])
plt.title("Average Happiness Score vs GDP per Capita (2016)")
plt.ylabel("Average Value")
plt.show()
```

```
data_2017 = annual_averages[annual_averages['Year'] == 2017]
data_2018 = annual_averages[annual_averages['Year'] == 2018]

h17 = data_2017['Happiness Score'].values[0]
g17 = data_2017['GDP per Capita'].values[0]

h18 = data_2018['Happiness Score'].values[0]
g18 = data_2018['GDP per Capita'].values[0]

plt.figure(figsize=(6,4))
plt.bar(["2017 H", "2017 GDP", "2018 H", "2018 GDP"], [h17, g17, h18, g18],
        color=["skyblue","salmon","skyblue","salmon"])
plt.title("Average Happiness Score vs GDP per Capita (2017 & 2018)")
plt.ylabel("Average Value")
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['Happiness Score'], kde=True, bins=20, color='blue')
plt.title('Overall Distribution of Happiness Score (2015-2018)')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.show()
```
